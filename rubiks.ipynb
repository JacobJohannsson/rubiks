```python
class Cube:
    
    def __init__(self):
        self.up = [['y','y','y'], ['y','y','y'], ['y','y','y']]
        self.down = [['w','w','w'], ['w','w','w'], ['w','w','w']]
        self.right = [['r','r','r'], ['r','r','r'], ['r','r','r']]
        self.left = [['o','o','o'], ['o','o','o'], ['o','o','o']]
        self.front = [['b','b','b'], ['b','b','b'], ['b','b','b']]
        self.back = [['g','g','g'], ['g','g','g'], ['g','g','g']]
        
    def print_cube(self):
        import sys
        
        colour_map = {
            'w': u"\u001b[48;5;" + '231' + "m" + ''.ljust(2),
            'y': u"\u001b[48;5;" + '226' + "m" + ''.ljust(2),
            'r': u"\u001b[48;5;" + '196' + "m" + ''.ljust(2),
            'b': u"\u001b[48;5;" + '27' + "m" + ''.ljust(2),
            'o': u"\u001b[48;5;" + '208' + "m" + ''.ljust(2),
            'g': u"\u001b[48;5;" + '46' + "m" + ''.ljust(2)
        }
        
        sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(39))
        print() 
        
        for row in self.up:
            sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(12))
            for colour in row:
                sys.stdout.write(colour_map[colour])
            sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(21))
            print()
        sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(39))
        print()

        for i in range(3):
            sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(3))
            counter = 0
            for colour in self.left[i] + self.front[i] + self.right[i] + self.back[i]:
                sys.stdout.write(colour_map[colour])
                counter += 1
                if counter % 3 == 0:
                    sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(3))
            print()
        sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(39))
        print()
        
        for row in self.down:
            sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(12))
            for colour in row:
                sys.stdout.write(colour_map[colour])
            sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(21))
            print()
        sys.stdout.write(u"\u001b[48;5;" + '232' + "m" + ''.ljust(39))
        print('\n\n')
    
    def R(self):
        self.right[0], self.right[1], self.right[2] = \
        [self.right[2][0], self.right[1][0], self.right[0][0]], \
        [self.right[2][1], self.right[1][1], self.right[0][1]], \
        [self.right[2][2], self.right[1][2], self.right[0][2]]
        
        self.front[0][2], self.front[1][2], self.front[2][2], self.up[0][2], self.up[1][2], self.up[2][2], \
        self.back[0][0], self.back[1][0], self.back[2][0], self.down[0][2], self.down[1][2], self.down[2][2] = \
        \
        self.down[0][2], self.down[1][2], self.down[2][2], self.front[0][2], self.front[1][2], self.front[2][2], \
        self.up[2][2], self.up[1][2], self.up[0][2], self.back[2][0], self.back[1][0], self.back[0][0]
       
    def R_prime(self):
        self.right[0], self.right[1], self.right[2] = \
        [self.right[0][2], self.right[1][2], self.right[2][2]], \
        [self.right[0][1], self.right[1][1], self.right[2][1]], \
        [self.right[0][0], self.right[1][0], self.right[2][0]]
       
        self.front[0][2], self.front[1][2], self.front[2][2], self.up[0][2], self.up[1][2], self.up[2][2], \
        self.back[2][0], self.back[1][0], self.back[0][0], self.down[0][2], self.down[1][2], self.down[2][2] = \
        \
        self.up[0][2], self.up[1][2], self.up[2][2], self.back[2][0], self.back[1][0], self.back[0][0], \
        self.down[0][2], self.down[1][2], self.down[2][2], self.front[0][2], self.front[1][2], self.front[2][2]
    
    def R2(self):
        self.R()
        self.R()
    
    def L(self):
        self.left[0], self.left[1], self.left[2] = \
        [self.left[2][0], self.left[1][0], self.left[0][0]], \
        [self.left[2][1], self.left[1][1], self.left[0][1]], \
        [self.left[2][2], self.left[1][2], self.left[0][2]]
        
        self.front[0][0], self.front[1][0], self.front[2][0], self.up[0][0], self.up[1][0], self.up[2][0], \
        self.back[2][2], self.back[1][2], self.back[0][2], self.down[0][0], self.down[1][0], self.down[2][0] = \
        \
        self.up[0][0], self.up[1][0], self.up[2][0], self.back[2][2], self.back[1][2], self.back[0][2], \
        self.down[0][0], self.down[1][0], self.down[2][0], self.front[0][0], self.front[1][0], self.front[2][0]
        
    def L_prime(self):
        self.left[0], self.left[1], self.left[2] = \
        [self.left[0][2], self.left[1][2], self.left[2][2]], \
        [self.left[0][1], self.left[1][1], self.left[2][1]], \
        [self.left[0][0], self.left[1][0], self.left[2][0]]
        
        self.front[0][0], self.front[1][0], self.front[2][0], self.up[0][0], self.up[1][0], self.up[2][0], \
        self.back[2][2], self.back[1][2], self.back[0][2], self.down[0][0], self.down[1][0], self.down[2][0] = \
        \
        self.down[0][0], self.down[1][0], self.down[2][0], self.front[0][0], self.front[1][0], self.front[2][0], \
        self.up[0][0], self.up[1][0], self.up[2][0], self.back[2][2], self.back[1][2], self.back[0][2]
    
    def L2(self):
        self.L()
        self.L()
    
    def F(self):
        self.front[0], self.front[1], self.front[2] = \
        [self.front[2][0], self.front[1][0], self.front[0][0]], \
        [self.front[2][1], self.front[1][1], self.front[0][1]], \
        [self.front[2][2], self.front[1][2], self.front[0][2]]
        
        self.left[0][2], self.left[1][2], self.left[2][2], self.up[2][0], self.up[2][1], self.up[2][2], \
        self.right[0][0], self.right[1][0], self.right[2][0], self.down[0][0], self.down[0][1], self.down[0][2] = \
        \
        self.down[0][0], self.down[0][1], self.down[0][2], self.left[2][2], self.left[1][2], self.left[0][2], \
        self.up[2][0], self.up[2][1], self.up[2][2], self.right[2][0], self.right[1][0], self.right[0][0]
    
    def F_prime(self):
        self.front[0], self.front[1], self.front[2] = \
        [self.front[0][2], self.front[1][2], self.front[2][2]], \
        [self.front[0][1], self.front[1][1], self.front[2][1]], \
        [self.front[0][0], self.front[1][0], self.front[2][0]]
        
        self.left[0][2], self.left[1][2], self.left[2][2], self.up[2][0], self.up[2][1], self.up[2][2], \
        self.right[0][0], self.right[1][0], self.right[2][0], self.down[0][0], self.down[0][1], self.down[0][2] = \
        \
        self.up[2][2], self.up[2][1], self.up[2][0], self.right[0][0], self.right[1][0], self.right[2][0], \
        self.down[0][2], self.down[0][1], self.down[0][0], self.left[0][2], self.left[1][2], self.left[2][2]
    
    def F2(self):
        self.F()
        self.F()
        
    def B(self):
        self.back[0], self.back[1], self.back[2] = \
        [self.back[2][0], self.back[1][0], self.back[0][0]], \
        [self.back[2][1], self.back[1][1], self.back[0][1]], \
        [self.back[2][2], self.back[1][2], self.back[0][2]]
        
        self.left[0][0], self.left[1][0], self.left[2][0], self.up[0][0], self.up[0][1], self.up[0][2], \
        self.right[0][2], self.right[1][2], self.right[2][2], self.down[2][2], self.down[2][1], self.down[2][0] = \
        \
        self.up[0][2], self.up[0][1], self.up[0][0], self.right[0][2], self.right[1][2], self.right[2][2], \
        self.down[2][2], self.down[2][1], self.down[2][0], self.left[2][0], self.left[1][0], self.left[0][0]
    
    def B_prime(self):
        self.back[0], self.back[1], self.back[2] = \
        [self.back[0][2], self.back[1][2], self.back[2][2]], \
        [self.back[0][1], self.back[1][1], self.back[2][1]], \
        [self.back[0][0], self.back[1][0], self.back[2][0]]
        
        self.left[0][0], self.left[1][0], self.left[2][0], self.up[0][0], self.up[0][1], self.up[0][2], \
        self.right[0][2], self.right[1][2], self.right[2][2], self.down[2][2], self.down[2][1], self.down[2][0] = \
        \
        self.down[2][0], self.down[2][1], self.down[2][2], self.left[2][0], self.left[1][0], self.left[0][0], \
        self.up[0][0], self.up[0][1], self.up[0][2], self.right[0][2], self.right[1][2], self.right[2][2]
    
    def B2(self):
        self.B()
        self.B()
        
    def U(self):
        self.up[0], self.up[1], self.up[2] = \
        [self.up[2][0], self.up[1][0], self.up[0][0]], \
        [self.up[2][1], self.up[1][1], self.up[0][1]], \
        [self.up[2][2], self.up[1][2], self.up[0][2]]
        
        self.front[0][0], self.front[0][1], self.front[0][2], self.right[0][0], self.right[0][1], self.right[0][2], \
        self.back[0][0], self.back[0][1], self.back[0][2], self.left[0][0], self.left[0][1], self.left[0][2] = \
        \
        self.right[0][0], self.right[0][1], self.right[0][2], self.back[0][0], self.back[0][1], self.back[0][2], \
        self.left[0][0], self.left[0][1], self.left[0][2], self.front[0][0], self.front[0][1], self.front[0][2]
        
    def U_prime(self):
        self.up[0], self.up[1], self.up[2] = \
        [self.up[0][2], self.up[1][2], self.up[2][2]], \
        [self.up[0][1], self.up[1][1], self.up[2][1]], \
        [self.up[0][0], self.up[1][0], self.up[2][0]]
        
        self.front[0][0], self.front[0][1], self.front[0][2], self.right[0][0], self.right[0][1], self.right[0][2], \
        self.back[0][0], self.back[0][1], self.back[0][2], self.left[0][0], self.left[0][1], self.left[0][2] = \
        \
        self.left[0][0], self.left[0][1], self.left[0][2], self.front[0][0], self.front[0][1], self.front[0][2], \
        self.right[0][0], self.right[0][1], self.right[0][2], self.back[0][0], self.back[0][1], self.back[0][2]
    
    def U2(self):
        self.U()
        self.U()
        
    def D(self):
        self.down[0], self.down[1], self.down[2] = \
        [self.down[2][0], self.down[1][0], self.down[0][0]], \
        [self.down[2][1], self.down[1][1], self.down[0][1]], \
        [self.down[2][2], self.down[1][2], self.down[0][2]]
        
        self.front[2][0], self.front[2][1], self.front[2][2], self.right[2][0], self.right[2][1], self.right[2][2], \
        self.back[2][0], self.back[2][1], self.back[2][2], self.left[2][0], self.left[2][1], self.left[2][2] = \
        \
        self.left[2][0], self.left[2][1], self.left[2][2], self.front[2][0], self.front[2][1], self.front[2][2], \
        self.right[2][0], self.right[2][1], self.right[2][2], self.back[2][0], self.back[2][1], self.back[2][2]
    
    def D_prime(self):
        self.down[0], self.down[1], self.down[2] = \
        [self.down[0][2], self.down[1][2], self.down[2][2]], \
        [self.down[0][1], self.down[1][1], self.down[2][1]], \
        [self.down[0][0], self.down[1][0], self.down[2][0]]
        
        self.front[2][0], self.front[2][1], self.front[2][2], self.right[2][0], self.right[2][1], self.right[2][2], \
        self.back[2][0], self.back[2][1], self.back[2][2], self.left[2][0], self.left[2][1], self.left[2][2] = \
        \
        self.right[2][0], self.right[2][1], self.right[2][2], self.back[2][0], self.back[2][1], self.back[2][2], \
        self.left[2][0], self.left[2][1], self.left[2][2], self.front[2][0], self.front[2][1], self.front[2][2]
    
    def D2(self):
        self.D()
        self.D()
        
    def x(self):
        self.down[0], self.down[1], self.down[2], self.front[0], self.front[1], self.front[2], \
        self.up[0], self.up[1], self.up[2], self.back[0], self.back[1], self.back[2] = \
        self.back[2], self.back[1], self.back[0], self.down[0], self.down[1], self.down[2], \
        self.front[0], self.front[1], self.front[2], self.up[2], self.up[1], self.up[0]
        
        self.down[0].reverse()
        self.down[1].reverse()
        self.down[2].reverse()
        self.back[0].reverse()
        self.back[1].reverse()
        self.back[2].reverse()
        
        self.right[0], self.right[1], self.right[2] = \
        [self.right[2][0], self.right[1][0], self.right[0][0]], \
        [self.right[2][1], self.right[1][1], self.right[0][1]], \
        [self.right[2][2], self.right[1][2], self.right[0][2]]
        
        self.left[0], self.left[1], self.left[2] = \
        [self.left[0][2], self.left[1][2], self.left[2][2]], \
        [self.left[0][1], self.left[1][1], self.left[2][1]], \
        [self.left[0][0], self.left[1][0], self.left[2][0]]
        
    def x_prime(self):
        self.down[0], self.down[1], self.down[2], self.front[0], self.front[1], self.front[2], \
        self.up[0], self.up[1], self.up[2], self.back[0], self.back[1], self.back[2] = \
        self.front[0], self.front[1], self.front[2], self.up[0], self.up[1], self.up[2], \
        self.back[2], self.back[1], self.back[0], self.down[2], self.down[1], self.down[0]
        
        self.up[0].reverse()
        self.up[1].reverse()
        self.up[2].reverse()
        self.back[0].reverse()
        self.back[1].reverse()
        self.back[2].reverse()
        
        self.right[0], self.right[1], self.right[2] = \
        [self.right[0][2], self.right[1][2], self.right[2][2]], \
        [self.right[0][1], self.right[1][1], self.right[2][1]], \
        [self.right[0][0], self.right[1][0], self.right[2][0]]
        
        self.left[0], self.left[1], self.left[2] = \
        [self.left[2][0], self.left[1][0], self.left[0][0]], \
        [self.left[2][1], self.left[1][1], self.left[0][1]], \
        [self.left[2][2], self.left[1][2], self.left[0][2]]  
        
    def x2(self):
        self.x()
        self.x()
        
    def y(self):
        self.front[0], self.front[1], self.front[2], self.left[0], self.left[1], self.left[2], \
        self.back[0], self.back[1], self.back[2], self.right[0], self.right[1], self.right[2] = \
        self.right[0], self.right[1], self.right[2], self.front[0], self.front[1], self.front[2], \
        self.left[0], self.left[1], self.left[2], self.back[0], self.back[1], self.back[2]
        
        self.up[0], self.up[1], self.up[2] = \
        [self.up[2][0], self.up[1][0], self.up[0][0]], \
        [self.up[2][1], self.up[1][1], self.up[0][1]], \
        [self.up[2][2], self.up[1][2], self.up[0][2]]
        
        self.down[0], self.down[1], self.down[2] = \
        [self.down[0][2], self.down[1][2], self.down[2][2]], \
        [self.down[0][1], self.down[1][1], self.down[2][1]], \
        [self.down[0][0], self.down[1][0], self.down[2][0]]
        
    def y_prime(self):
        self.front[0], self.front[1], self.front[2], self.left[0], self.left[1], self.left[2], \
        self.back[0], self.back[1], self.back[2], self.right[0], self.right[1], self.right[2] = \
        self.left[0], self.left[1], self.left[2], self.back[0], self.back[1], self.back[2], \
        self.right[0], self.right[1], self.right[2], self.front[0], self.front[1], self.front[2]
        
        self.up[0], self.up[1], self.up[2] = \
        [self.up[0][2], self.up[1][2], self.up[2][2]], \
        [self.up[0][1], self.up[1][1], self.up[2][1]], \
        [self.up[0][0], self.up[1][0], self.up[2][0]]
        
        self.down[0], self.down[1], self.down[2] = \
        [self.down[2][0], self.down[1][0], self.down[0][0]], \
        [self.down[2][1], self.down[1][1], self.down[0][1]], \
        [self.down[2][2], self.down[1][2], self.down[0][2]]
        
    def y2(self):
        self.y()
        self.y()
        
    def z(self):
        self.down[2], self.down[1], self.down[0], self.left[2], self.left[1], self.left[0], \
        self.up[2], self.up[1], self.up[0], self.right[2], self.right[1], self.right[0] = \
        [self.right[2][2], self.right[1][2], self.right[0][2]], [self.right[2][1], self.right[1][1], self.right[0][1]], \
        [self.right[2][0], self.right[1][0], self.right[0][0]], [self.down[2][2], self.down[1][2], self.down[0][2]], \
        [self.down[2][1], self.down[1][1], self.down[0][1]], [self.down[2][0], self.down[1][0], self.down[0][0]], \
        [self.left[2][2], self.left[1][2], self.left[0][2]], [self.left[2][1], self.left[1][1], self.left[0][1]], \
        [self.left[2][0], self.left[1][0], self.left[0][0]], [self.up[2][2], self.up[1][2], self.up[0][2]], \
        [self.up[2][1], self.up[1][1], self.up[0][1]], [self.up[2][0], self.up[1][0], self.up[0][0]]
        
        self.front[0], self.front[1], self.front[2] = \
        [self.front[2][0], self.front[1][0], self.front[0][0]], \
        [self.front[2][1], self.front[1][1], self.front[0][1]], \
        [self.front[2][2], self.front[1][2], self.front[0][2]]
        
        self.back[0], self.back[1], self.back[2] = \
        [self.back[0][2], self.back[1][2], self.back[2][2]], \
        [self.back[0][1], self.back[1][1], self.back[2][1]], \
        [self.back[0][0], self.back[1][0], self.back[2][0]]
                
    def z_prime(self):
        self.up[0], self.up[1], self.up[2], self.right[0], self.right[1], self.right[2], \
        self.down[0], self.down[1], self.down[2], self.left[0], self.left[1], self.left[2] = \
        [self.right[0][2], self.right[1][2], self.right[2][2]], [self.right[0][1], self.right[1][1], self.right[2][1]], \
        [self.right[0][0], self.right[1][0], self.right[2][0]], [self.down[0][2], self.down[1][2], self.down[2][2]], \
        [self.down[0][1], self.down[1][1], self.down[2][1]], [self.down[0][0], self.down[1][0], self.down[2][0]], \
        [self.left[0][2], self.left[1][2], self.left[2][2]], [self.left[0][1], self.left[1][1], self.left[2][1]], \
        [self.left[0][0], self.left[1][0], self.left[2][0]], [self.up[0][2], self.up[1][2], self.up[2][2]], \
        [self.up[0][1], self.up[1][1], self.up[2][1]], [self.up[0][0], self.up[1][0], self.up[2][0]]
        
        self.front[0], self.front[1], self.front[2] = \
        [self.front[0][2], self.front[1][2], self.front[2][2]], \
        [self.front[0][1], self.front[1][1], self.front[2][1]], \
        [self.front[0][0], self.front[1][0], self.front[2][0]]
        
        self.back[0], self.back[1], self.back[2] = \
        [self.back[2][0], self.back[1][0], self.back[0][0]], \
        [self.back[2][1], self.back[1][1], self.back[0][1]], \
        [self.back[2][2], self.back[1][2], self.back[0][2]]
        
    def z2(self):
        self.z()
        self.z()
        
    def r(self):
        self.L()
        self.x()
        
    def r_prime(self):
        self.L_prime()
        self.x_prime()
        
    def r2(self):
        self.r()
        self.r()
    
    def l(self):
        self.R()
        self.x_prime()
    
    def l_prime(self):
        self.R_prime()
        self.x()        
        
    def l2(self):
        self.l()
        self.l()
        
    def f(self):
        self.B()
        self.z()
    
    def f_prime(self):
        self.B_prime()
        self.z_prime()
    
    def f2(self):
        self.f()
        self.f()
    
    def b(self):
        self.F()
        self.z_prime()
    
    def b_prime(self):
        self.F_prime()
        self.z()
        
    def b2(self):
        self.b()
        self.b()
    
    def u(self):
        self.D_prime()
        self.y()

    def u_prime(self):
        self.D()
        self.y_prime()
        
    def u2(self):
        self.u()
        self.u()
    
    def d(self):
        self.U()
        self.y_prime()
    
    def d_prime(self):
        self.U_prime()
        self.y()
        
    def d2(self):
        self.d()
        self.d()
        
    def M(self):
        self.R()
        self.L_prime()
        self.x_prime()
        
    def M_prime(self):
        self.R_prime()
        self.L()
        self.x()
    
    def M2(self):
        self.M()
        self.M()
        
    def E(self):
        self.U()
        self.D_prime()
        self.y_prime()
    
    def E_prime(self):
        self.U_prime()
        self.D()
        self.y()
    
    def E2(self):
        self.E()
        self.E()
    
    def S(self):
        self.F_prime()
        self.B()
        self.z()
    
    def S_prime(self):
        self.F()
        self.B_prime()
        self.z_prime()
        
    def S2(self):
        self.S()
        self.S()
    
    def scramble(self):
        import random
        scramble_alg = ""
        move_list = ["R", "R'", "L", "L'", "U", "U'", "D", "D'", "F", "F'", "B", "B'", "R2", "L2", "U2", "D2", "F2", "B2"]
        for i in range(40):
            scramble_alg += (move_list[random.randrange(0,len(move_list))] + " ")
        self.perform_algorithm(scramble_alg)
        return scramble_alg
    
    def reset(self):
        self.up = [['y','y','y'], ['y','y','y'], ['y','y','y']]
        self.down = [['w','w','w'], ['w','w','w'], ['w','w','w']]
        self.right = [['r','r','r'], ['r','r','r'], ['r','r','r']]
        self.left = [['o','o','o'], ['o','o','o'], ['o','o','o']]
        self.front = [['b','b','b'], ['b','b','b'], ['b','b','b']]
        self.back = [['g','g','g'], ['g','g','g'], ['g','g','g']]
        cube.print_cube()    
    
    def perform_algorithm(self, alg):
        move_index = {
            "R": Cube.R,
            "R'": Cube.R_prime,
            "L": Cube.L,
            "L'": Cube.L_prime, 
            "U": Cube.U,
            "U'": Cube.U_prime,
            "D": Cube.D,
            "D'": Cube.D_prime,
            "F": Cube.F,
            "F'": Cube.F_prime,
            "B": Cube.B,
            "B'": Cube.B_prime,
            "R2": Cube.R2,
            "L2": Cube.L2,
            "U2": Cube.U2,
            "D2": Cube.D2,
            "F2": Cube.F2,
            "B2": Cube.B2,
            "x": Cube.x,
            "x'": Cube.x_prime,
            "y": Cube.y,
            "y'": Cube.y_prime,
            "z": Cube.z,
            "z'": Cube.z_prime,
            "x2": Cube.x2,
            "y2": Cube.y2,
            "z2": Cube.z2,
            "r": Cube.r,
            "r'": Cube.r_prime,
            "l": Cube.l,
            "l'": Cube.l_prime,
            "u": Cube.u,
            "u'": Cube.u_prime,
            "d": Cube.d,
            "d'": Cube.d_prime,
            "f": Cube.f,
            "f'": Cube.f_prime,
            "r2": Cube.r2,
            "l2": Cube.l2,
            "u2": Cube.u2,
            "d2": Cube.d2,
            "f2": Cube.f2,
            "b2": Cube.b2,
            "M": Cube.M,
            "M'": Cube.M_prime,
            "M2": Cube.M2,
            "E": Cube.E,
            "E'": Cube.E_prime,
            "E2": Cube.E2,
            "S": Cube.S,
            "S'": Cube.S_prime,
            "S2": Cube.S2
        }
        alg = alg.split()
        for move in alg:
            move_index[move](self)
        
    def moves_to_solve(self, alg):
        move_index = {
            "R": Cube.R,
            "R'": Cube.R_prime,
            "L": Cube.L,
            "L'": Cube.L_prime, 
            "U": Cube.U,
            "U'": Cube.U_prime,
            "D": Cube.D,
            "D'": Cube.D_prime,
            "F": Cube.F,
            "F'": Cube.F_prime,
            "B": Cube.B,
            "B'": Cube.B_prime,
            "R2": Cube.R2,
            "L2": Cube.L2,
            "U2": Cube.U2,
            "D2": Cube.D2,
            "F2": Cube.F2,
            "B2": Cube.B2,
            "x": Cube.x,
            "x'": Cube.x_prime,
            "y": Cube.y,
            "y'": Cube.y_prime,
            "z": Cube.z,
            "z'": Cube.z_prime,
            "x2": Cube.x2,
            "y2": Cube.y2,
            "z2": Cube.z2,
            "r": Cube.r,
            "r'": Cube.r_prime,
            "l": Cube.l,
            "l'": Cube.l_prime,
            "u": Cube.u,
            "u'": Cube.u_prime,
            "d": Cube.d,
            "d'": Cube.d_prime,
            "f": Cube.f,
            "f'": Cube.f_prime,
            "r2": Cube.r2,
            "l2": Cube.l2,
            "u2": Cube.u2,
            "d2": Cube.d2,
            "f2": Cube.f2,
            "b2": Cube.b2,
            "M": Cube.M,
            "M'": Cube.M_prime,
            "M2": Cube.M2,
            "E": Cube.E,
            "E'": Cube.E_prime,
            "E2": Cube.E2,
            "S": Cube.S,
            "S'": Cube.S_prime,
            "S2": Cube.S2
        }
        alg = alg.split()
        counter = 0
        while not self.is_solved() or counter == 0:
            for move in alg:
                move_index[move](self)
            counter += 1
        return counter
        
    def is_empty_cross_slot(self, coords):
        return (self.down[coords[0]][coords[1]] != self.down[1][1])
        
    def cross(self):
        return (self.down[1][1] == self.down[0][1] == self.down[1][0] == self.down[1][2] == self.down[2][1])
    
    def solve_cross(self):
        moves = ""
        while not self.cross():
            side_list = [self.up, self.front, self.back, self.right, self.left]
            for i in range(len(side_list)):
                edge_dict = {0: [0,1], 1: [1,0], 2: [1,2], 3: [2,1]}
                for j in edge_dict:
                    if side_list[i][edge_dict[j][0]][edge_dict[j][1]] == self.down[1][1]:
                        if i == 0:
                            edge_switch = {(2,1): [0,1], (1,2): [1,2], (1,0): [1,0], (0,1): [2,1]}
                            while not self.is_empty_cross_slot(edge_switch[tuple(edge_dict[j])]):
                                self.perform_algorithm("D ")
                                moves += "D "
                            move_dict = {(2,1): "F2 ", (1,2): "R2 ", (1,0): "L2 ", (0,1): "B2 "}
                            self.perform_algorithm(move_dict[tuple(edge_dict[j])])
                            moves += move_dict[tuple(edge_dict[j])]
                        else:
                            if i == 1:
                                if j == 0:
                                    while not self.is_empty_cross_slot([1,0]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("F' L F ")
                                    moves += "F' L F "
                                if j == 1:
                                    while not self.is_empty_cross_slot([1,0]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("L ")
                                    moves += "L "
                                if j == 2:
                                    while not self.is_empty_cross_slot([1,2]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("R' ")
                                    moves += "R' "
                            elif i == 2:
                                if j == 0:
                                    while not self.is_empty_cross_slot([1,0]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("B L' B' ")
                                    moves += "B L' B' "
                                if j == 1:
                                    while not self.is_empty_cross_slot([1,0]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("R ")
                                    moves += "R "
                                if j == 2:
                                    while not self.is_empty_cross_slot([1,2]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("L' ")
                                    moves += "L' "
                            elif i == 3:
                                if j == 0:
                                    while not self.is_empty_cross_slot([0,1]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("R' F R ")
                                    moves += "R' F R "
                                if j == 1:
                                    while not self.is_empty_cross_slot([0,1]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("F ")
                                    moves += "F "
                                if j == 2:
                                    while not self.is_empty_cross_slot([2,1]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("B' ")
                                    moves += "B' "
                            elif i == 4:
                                if j == 0:
                                    while not self.is_empty_cross_slot([0,1]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("L F' L' ")
                                    moves += "L F' L' "
                                if j == 1:
                                    while not self.is_empty_cross_slot([2,1]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("B ")
                                    moves += "B "
                                if j == 2:
                                    while not self.is_empty_cross_slot([0,1]):
                                        self.perform_algorithm("D ")
                                        moves += "D "
                                    self.perform_algorithm("F' ")
                                    moves += "F' "
                            if j == 3:
                                move_dict = {1: "F2 ", 3: "R2 ", 4: "L2 ", 2: "B2 "}
                                self.perform_algorithm(move_dict[i])
                                moves += move_dict[i]
        return moves
        
    def cross_sides(self):
        return ((self.left[1][1] == self.left[2][1]) and (self.front[1][1] == self.front[2][1]) and \
        (self.right[1][1] == self.right[2][1]) and (self.back[1][1] == self.back[2][1]))
    
    def solve_cross_sides(self):
        moves = ""
        while not self.back[2][1] == self.back[1][1]:
            self.perform_algorithm("D ")
            moves += "D "
        if self.left[2][1] == self.front[1][1]:
            self.perform_algorithm("L2 U' F2 U L2 ")
            moves += "L2 U' F2 U L2 "
        if self.left[2][1] == self.right[1][1]:
            self.perform_algorithm("L2 U2 R2 U2 L2 ")
            moves += "L2 U2 R2 U2 L2 "
        if self.left[2][1] == self.front[1][1]:
            self.perform_algorithm("L2 U' F2 U L2 ")
            moves += "L2 U' F2 U L2 "
        if self.front[2][1] == self.right[1][1]:
            self.perform_algorithm("F2 U' R2 U F2 ")
            moves += "F2 U' R2 U F2 "
        return moves
            
    def is_empty_corner_slot(self, coords):
        return (self.down[coords[0]][coords[1]] != self.down[1][1])
    
    def corners(self):
        return (self.down[0][0] == self.down[0][2] == self.down[2][0] == self.down[2][2] == self.down[1][1])
    
    def solve_corners(self):
        moves = ""
        while not self.corners():
            side_list = [self.up, self.front, self.back, self.right, self.left]
            for i in range(len(side_list)):
                corner_dict = {0: [0,0], 1: [2,0], 2: [0,2], 3: [2,2]}
                for j in corner_dict:
                    if side_list[i][corner_dict[j][0]][corner_dict[j][1]] == self.down[1][1]:
                        if i == 0:
                            corner_switch = {0: [2,0], 2: [2,2], 1: [0,0], 3: [0,2]}
                            while not self.is_empty_corner_slot(corner_switch[j]):
                                self.perform_algorithm("D ")
                                moves += "D "
                            move_dict = {0: "y2 " + 3*"R U R' U' " + "y2 ", 2: "y " + 3*"R U R' U' " + "y' ", \
                                         1: "y' " + 3*"R U R' U' " + "y ", 3: 3*"R U R' U' "}
                            self.perform_algorithm(move_dict[j])
                            moves += move_dict[j]
                        else:
                            if j == 2:
                                corresponding_corner_dict = {1: [0,2], 2: [2,0], 3: [2,2], 4: [0,0]}
                                while not self.is_empty_corner_slot(corresponding_corner_dict[i]):
                                    self.perform_algorithm("D ")
                                    moves += "D "
                                move_dict = {1: "R U' R' ", 2: "y2 R U' R' y2 ", \
                                             3: "y R U' R' y' ", 4: "y' R U' R' y "}
                                self.perform_algorithm(move_dict[i])
                                moves += move_dict[i]
                            elif j == 0:
                                corresponding_corner_dict = {1: [0,0], 2: [2,2], 3: [0,2], 4: [2,0]}
                                while not self.is_empty_corner_slot(corresponding_corner_dict[i]):
                                    self.perform_algorithm("D ")
                                    moves += "D "
                                move_dict = {1: "L' U L ", 2: "y2 L' U L y2 ", \
                                             3: "y L' U L y' ", 4: "y' L' U L y "}
                                self.perform_algorithm(move_dict[i])
                                moves += move_dict[i]
                            elif j == 1:
                                move_dict = {1: "L' U L ", 2: "y2 L' U L y2 ", \
                                             3: "y L' U L y' ", 4: "y' L' U L y "}
                                self.perform_algorithm(move_dict[i])
                                moves += move_dict[i]
                            elif j == 3:
                                move_dict = {1: "R U' R' ", 2: "y2 R U' R' y2 ", \
                                             3: "y R U' R' y' ", 4: "y' R U' R' y "}
                                self.perform_algorithm(move_dict[i])
                                moves += move_dict[i]
        while self.front[1][1] != self.front[2][1]:
            self.perform_algorithm("D ")
            moves += "D "
        return moves
            
    def corner_sides(self):
        return (self.front[2][0] == self.front[2][2] == self.front[1][1]) and (self.left[2][0] == self.left[2][2] == \
                self.left[1][1]) and (self.right[2][0] == self.right[2][2] == self.right[1][1]) and (self.back[2][0] == \
                self.back[2][2] == self.back[1][1])
        
    def solve_corner_sides(self):
        moves = ""
        if self.front[2][0] == self.left[1][1]:
            self.perform_algorithm("y' R U R' L' U' L R U R' y ")
            moves += "y' R U R' L' U' L R U R' y "
        elif self.right[2][0] == self.left[1][1]:
            self.perform_algorithm("R U R' U2 L U' L' U2 R U R' ")
            moves += "R U R' U2 L U' L' U2 R U R' "
        elif self.back[2][0] == self.left[1][1]:
            self.perform_algorithm("y2 R U R' L' U' L R U R' y2 ")
            moves += "y2 R U R' L' U' L R U R' y2 "
        if self.back[2][0] == self.front[1][1]:
            self.perform_algorithm("R' U' R U2 L' U L U2 R' U' R ")
            moves += "R' U' R U2 L' U L U2 R' U' R "
        elif self.right[2][0] == self.front[1][1]:
            self.perform_algorithm("R U R' L' U' L R U R' ")
            moves += "R U R' L' U' L R U R' "
        if self.back[2][0] == self.right[1][1]:
            self.perform_algorithm("y R U R' L' U' L R U R' y' ")
            moves += "y R U R' L' U' L R U R' y' "
        return moves
    
    def f2l_edges(self):
        return (self.front[1][0] == self.front[1][1] == self.front[1][2]) and (self.right[1][0] == self.right[1][1] == \
            self.right[1][2]) and (self.back[1][0] == self.back[1][1] == self.back [1][2]) and (self.left[1][0] == \
            self.left[1][1] == self.left[1][2])
            
    
    def solve_f2l_edges(self):
        moves = ""
        possible_edges = {
            0: {self.up[0][1],self.back[0][1]},
            1: {self.up[1][2],self.right[0][1]},
            2: {self.up[2][1],self.front[0][1]},
            3: {self.up[1][0],self.left[0][1]},
            4: {self.front[1][0],self.left[1][2]},
            5: {self.right[1][0],self.front[1][2]},
            6: {self.back[1][0],self.right[1][2]},
            7: {self.left[1][0],self.back[1][2]},
        }
        for i in range(8):
            if possible_edges[i] == {self.back[1][1], self.left[1][1]}:
                if i in [0,1,2,3]:
                    if i == 0:
                        self.perform_algorithm("U' ")
                        moves += "U' "
                    elif i == 1:
                        self.perform_algorithm("U2 ")
                        moves += "U2 "
                    elif i == 2:
                        self.perform_algorithm("U ")
                        moves += "U "
                    if self.left[1][1] != self.left[0][1]:
                        self.perform_algorithm("y2 U2 R U' R' y U' L' U L y ")
                        moves += "y2 U2 R U' R' y U' L' U L y "
                    else:
                        self.perform_algorithm("y' U' L' U L y' U R U' R' y2 ")
                        moves += "y' U' L' U L y' U R U' R' y2 "
                elif i in [4,5,6]:
                    if i == 4:
                        self.perform_algorithm("L' U L y' U R U' R' y U' ")
                        moves += "L' U L y' U R U' R' y U' "
                    elif i == 5:
                        self.perform_algorithm("R U' R' y U' L' U L y' U' ")
                        moves += "R U' R' y U' L' U L y' U' "
                    elif i == 6:
                        self.perform_algorithm("y R U' R' y U' L' U L y2 ")
                        moves += "y R U' R' y U' L' U L y2 "
                    if self.left[1][1] != self.left[0][1]:
                        self.perform_algorithm("y2 U2 R U' R' y U' L' U L y ")
                        moves += "y2 U2 R U' R' y U' L' U L y "
                    else:
                        self.perform_algorithm("y' U' L' U L y' U R U' R' y2 ")
                        moves += "y' U' L' U L y' U R U' R' y2 "
                elif i == 7:
                    if self.left[1][1] != self.left[1][0]:
                        self.perform_algorithm("y2 R U' R' y U' L' U L y' U' R U' R' y U' L' U L y ")
                        moves += "y2 R U' R' y U' L' U L y' U' R U' R' y U' L' U L y "
        possible_edges = {
            0: {self.up[0][1],self.back[0][1]},
            1: {self.up[1][2],self.right[0][1]},
            2: {self.up[2][1],self.front[0][1]},
            3: {self.up[1][0],self.left[0][1]},
            4: {self.front[1][0],self.left[1][2]},
            5: {self.right[1][0],self.front[1][2]},
            6: {self.back[1][0],self.right[1][2]},
            7: {self.left[1][0],self.back[1][2]},
        }
        for i in range(7):
            if possible_edges[i] == {self.back[1][1], self.right[1][1]}:
                if i in [0,1,2,3]:
                    if i == 1:
                        self.perform_algorithm("U' ")
                        moves += "U' "
                    elif i == 2:
                        self.perform_algorithm("U2 ")
                        moves += "U2 "
                    elif i == 3:
                        self.perform_algorithm("U ")
                        moves += "U "
                    if self.back[1][1] != self.back[0][1]:
                        self.perform_algorithm("y U2 R U' R' y U' L' U L y2 ")
                        moves += "y U2 R U' R' y U' L' U L y2 "
                    else:
                        self.perform_algorithm("y2 U' L' U L y' U R U' R' y' ")
                        moves += "y2 U' L' U L y' U R U' R' y' "
                elif i in [4,5]:
                    if i == 4:
                        self.perform_algorithm("L' U L y' U R U' R' y ")
                        moves += "L' U L y' U R U' R' y "
                    elif i == 5:
                        self.perform_algorithm("R U' R' y U' L' U L y' ")
                        moves += "R U' R' y U' L' U L y' "
                    if self.back[1][1] != self.back[0][1]:
                        self.perform_algorithm("y U2 R U' R' y U' L' U L y2 ")
                        moves += "y U2 R U' R' y U' L' U L y2 "
                    else:
                        self.perform_algorithm("y2 U' L' U L y' U R U' R' y' ")
                        moves += "y2 U' L' U L y' U R U' R' y' "
                elif i == 6:
                    if self.back[1][1] != self.back[1][0]:
                        self.perform_algorithm("y R U' R' y U' L' U L y' U' R U' R' y U' L' U L y2 ")
                        moves += "y R U' R' y U' L' U L y' U' R U' R' y U' L' U L y2 "
        possible_edges = {
            0: {self.up[0][1],self.back[0][1]},
            1: {self.up[1][2],self.right[0][1]},
            2: {self.up[2][1],self.front[0][1]},
            3: {self.up[1][0],self.left[0][1]},
            4: {self.front[1][0],self.left[1][2]},
            5: {self.right[1][0],self.front[1][2]},
            6: {self.back[1][0],self.right[1][2]},
            7: {self.left[1][0],self.back[1][2]},
        }
        for i in range(6):
            if possible_edges[i] == {self.front[1][1], self.right[1][1]}:
                if i in [0,1,2,3]:
                    if i == 0:
                        self.perform_algorithm("U ")
                        moves += "U "
                    elif i == 2:
                        self.perform_algorithm("U' ")
                        moves += "U' "
                    elif i == 3:
                        self.perform_algorithm("U2 ")
                        moves += "U2 "
                    if self.right[1][1] != self.right[0][1]:
                        self.perform_algorithm("U2 R U' R' y U' L' U L y' ")
                        moves += "U2 R U' R' y U' L' U L y' "
                    else:
                        self.perform_algorithm("y U' L' U L y' U R U' R' ")
                        moves += "y U' L' U L y' U R U' R' "
                elif i == 4:
                    self.perform_algorithm("L' U L y' U R U' R' y U ")
                    moves += "L' U L y' U R U' R' y U "
                    if self.right[1][1] != self.right[0][1]:
                        self.perform_algorithm("U2 R U' R' y U' L' U L y' ")
                        moves += "U2 R U' R' y U' L' U L y' "
                    else:
                        self.perform_algorithm("y U' L' U L y' U R U' R' ")
                        moves += "y U' L' U L y' U R U' R' "
                elif i == 5:
                    if self.right[1][1] != self.right[1][0]:
                        self.perform_algorithm("R U' R' y U' L' U L y' U' R U' R' y U' L' U L y' ")
                        moves += "R U' R' y U' L' U L y' U' R U' R' y U' L' U L y' "
        possible_edges = {
            0: {self.up[0][1],self.back[0][1]},
            1: {self.up[1][2],self.right[0][1]},
            2: {self.up[2][1],self.front[0][1]},
            3: {self.up[1][0],self.left[0][1]},
            4: {self.front[1][0],self.left[1][2]},
            5: {self.right[1][0],self.front[1][2]},
            6: {self.back[1][0],self.right[1][2]},
            7: {self.left[1][0],self.back[1][2]},
        }
        for i in range(5):
            if possible_edges[i] == {self.front[1][1], self.left[1][1]}:
                if i in [0,1,2,3]:
                    if i == 0:
                        self.perform_algorithm("U2 ")
                        moves += "U2 "
                    elif i == 1:
                        self.perform_algorithm("U ")
                        moves += "U "
                    elif i == 3:
                        self.perform_algorithm("U' ")
                        moves += "U' "
                    if self.front[1][1] != self.front[0][1]:
                        self.perform_algorithm("y' U2 R U' R' y U' L' U L ")
                        moves += "y' U2 R U' R' y U' L' U L "
                    else:
                        self.perform_algorithm("U' L' U L y' U R U' R' y ")
                        moves += "U' L' U L y' U R U' R' y "
                elif i == 4:
                    if self.front[1][1] != self.front[1][0]:
                        self.perform_algorithm("y' R U' R' y U' L' U L y' U' R U' R' y U' L' U L ")
                        moves += "y' R U' R' y U' L' U L y' U' R U' R' y U' L' U L "
        return moves
    
    def ll_cross(self):
        return [self.up[0][1], self.up[1][0], self.up[2][1], self.up[1][2]].count(self.up[1][1]) == 4
    
    def solve_ll_cross(self):
        moves = ""
        if [self.up[0][1], self.up[1][0], self.up[2][1], self.up[1][2]].count(self.up[1][1]) == 0:
            self.perform_algorithm("F R U R' U' F' f R U R' U' f' ")
            moves += "F R U R' U' F' f R U R' U' f' "
        elif [self.up[0][1], self.up[1][0], self.up[2][1], self.up[1][2]].count(self.up[1][1]) == 2:
            while not (self.up[2][1] != self.up[1][1] and self.up[1][0] == self.up[1][1]):
                self.perform_algorithm("U ")
                moves += "U "
            if self.up[1][2] == self.up[1][1]:
                self.perform_algorithm("F R U R' U' F' ")
                moves += "F R U R' U' F' "
            else:
                self.perform_algorithm("F U R U' R' F' ")
                moves += "F U R U' R' F' "
        return moves
                
    def is_top(self, sticker):
        return sticker == self.up[1][1]
                    
    def ll_corners(self):
        return [self.up[0][0], self.up[2][0], self.up[0][2], self.up[2][2]].count(self.up[1][1]) == 4
    
    def alg_21(self):
        return self.ll_cross() and not (self.is_top(self.up[0][0]) or self.is_top(self.up[0][2]) or \
        self.is_top(self.up[2][0]) or self.is_top(self.up[2][2])) and self.is_top(self.left[0][2]) \
        and self.is_top(self.left[0][0]) and self.is_top(self.right[0][0])
    
    def alg_22(self):
        return self.ll_cross() and not (self.is_top(self.up[0][0]) or self.is_top(self.up[0][2]) or \
        self.is_top(self.up[2][0]) or self.is_top(self.up[2][2])) and self.is_top(self.front[0][2]) \
        and self.is_top(self.left[0][2])
    
    def alg_23(self):
        return self.ll_cross() and self.is_top(self.up[0][0]) and self.is_top(self.up[0][2]) and \
        not (self.is_top(self.up[2][0]) and self.is_top(self.up[2][2])) and self.is_top(self.front[0][0])
    
    def alg_24(self):
        return self.ll_cross() and self.is_top(self.up[0][2]) and self.is_top(self.up[2][2]) and \
        not (self.is_top(self.up[0][0]) or self.is_top(self.up[2][0])) and self.is_top(self.front[0][0])
    
    def alg_25(self):
        return self.ll_cross() and self.is_top(self.up[0][0]) and self.is_top(self.up[2][2]) and \
        not (self.is_top(self.up[2][0]) or self.is_top(self.up[0][2])) and self.is_top(self.front[0][0])
    
    def alg_26(self):
        return self.ll_cross() and self.is_top(self.up[0][0]) and not (self.is_top(self.up[2][0]) or \
        self.is_top(self.up[0][2]) or self.is_top(self.up[2][2])) and self.is_top(self.back[0][0])
    
    def alg_27(self):
        return self.ll_cross() and self.is_top(self.up[2][0]) and not (self.is_top(self.up[0][0]) or \
        self.is_top(self.up[0][2]) or self.is_top(self.up[2][2])) and self.is_top(self.front[0][2])
    
    def solve_ll_corners(self):
        moves = ""
        while not (self.alg_21() or self.alg_22() or self.alg_23() or self.alg_24() or self.alg_25() or \
                   self.alg_26() or self.alg_27() or self.ll_corners()):
            self.perform_algorithm("U ")
            moves += "U "
        if self.alg_21():
            self.perform_algorithm("R U R' U R U' R' U R U2 R' ")
            moves += "R U R' U R U' R' U R U2 R' "
        elif self.alg_22():
            self.perform_algorithm("R U2 R2 U' R2 U' R2 U2 R ")
            moves += "R U2 R2 U' R2 U' R2 U2 R "
        elif self.alg_23():
            self.perform_algorithm("R2 D R' U2 R D' R' U2 R' ")
            moves += "R2 D R' U2 R D' R' U2 R' "
        elif self.alg_24():
            self.perform_algorithm("r U R' U' r' F R F' ")
            moves += "r U R' U' r' F R F' "
        elif self.alg_25():
            self.perform_algorithm("F R' F' r U R U' r' ")
            moves += "F R' F' r U R U' r' "
        elif self.alg_26():
            self.perform_algorithm("R' U' R U' R' U2 R ")
            moves += "R' U' R U' R' U2 R "
        elif self.alg_27():
            self.perform_algorithm("R U R' U R U2 R' ")
            moves += "R U R' U R U2 R' "
        return moves
    
    def ll_corner_sides(self):
        return self.front[0][0] == self.front[1][1] and self.left[0][0] == self.left[1][1] and \
        self.back[0][0] == self.back[1][1] and self.right[0][0] == self.right[1][1]
    
    def solve_ll_corner_sides(self):
        moves = ""
        while not self.front[0][0] == self.front[1][1]:
            self.perform_algorithm("U ")
            moves += "U "
        if self.front[0][2] == self.back[1][1] and self.left[0][0] == self.right[1][1]:
            self.perform_algorithm("F R U' R' U' R U R' F' R U R' U' R' F R F' ")
            moves += "F R U' R' U' R U R' F' R U R' U' R' F R F' "
        elif self.left[0][0] == self.left[1][1] and self.front[0][2] != self.front[1][1]:
            self.perform_algorithm("R U R' U' R' F R2 U' R' U' R U R' F' ")
            moves += "R U R' U' R' F R2 U' R' U' R U R' F' "
        elif self.front[0][2] == self.back[1][1] and self.left[0][0] != self.right[1][1]:
            self.perform_algorithm("x R' U  R' D2 R U' R' D2 R2 x' ")
            moves += "x R' U  R' D2 R U' R' D2 R2 x' "
        elif self.front[0][2] != self.back[1][1] and self.left[0][0] == self.right[1][1]:
            self.perform_algorithm("x R2 D2 R U R' D2 R U' R x' ")
            moves += "x R2 D2 R U R' D2 R U' R x' "
        elif self.front[0][2] == self.front[1][1] and self.left[0][0] != self.left[1][1]:
            self.perform_algorithm("U R U R' U' R' F R2 U' R' U' R U R' F' U' ")
            moves += "U R U R' U' R' F R2 U' R' U' R U R' F' U' "
        return moves
    
    def is_solved(self):
        solved = True
        faces = [self.up, self.down, self.right, self.left, self.front, self.back]
        for face in faces:
            colour = face[0][0]
            for row in face:
                for piece in row:
                    if piece != colour:
                        solved = False
        return solved
    
    def solve_ll_edges(self):
        moves = ""
        if not self.is_solved():
            if [self.front[0][1] == self.front[1][1], self.left[0][1] == self.left[1][1], \
                self.right[0][1] == self.right[1][1], self.back[0][1] == self.back[1][1]].count(True) == 0:
                self.perform_algorithm("M2 U' M U2 M' U' M2 ")
                moves += "M2 U' M U2 M' U' M2 "
            while self.back[0][1] != self.back[1][1]:
                self.perform_algorithm("y ")
                moves += "y "
            if self.left[0][1] == self.right[1][1]:
                self.perform_algorithm("M2 U' M U2 M' U' M2 ")
                moves += "M2 U' M U2 M' U' M2 "
            else:
                self.perform_algorithm("M2 U M U2 M' U M2 ")
                moves += "M2 U M U2 M' U M2 "
        return moves
    
    def full_solve(self):
        solution = ""
        solution += self.solve_cross()
        solution += self.solve_cross_sides()
        solution += self.solve_corners()
        solution += self.solve_corner_sides()
        solution += self.solve_f2l_edges()
        solution += self.solve_ll_cross()
        solution += self.solve_ll_corners()
        solution += self.solve_ll_corner_sides()
        solution += self.solve_ll_edges()
        return solution
    
def practice_solve():
    cube = Cube()
    scramble = cube.scramble()
    solution = cube.full_solve()
    print("Scramble:" + "\n" + scramble + "\n\n" + str(len(solution.split())) + " move solution:" + "\n" + solution)
    cube.perform_algorithm(scramble)
    cube.print_cube()
    cube.perform_algorithm(solution)
    cube.print_cube()        
```
